### Handling missing and outlier data

* Identify missing and outlier data and update it with mean for the columns containing numeric values. 

````python
def identify_outlier(sigma = 3.0):
    """
    Aims:
     - assuming each attribute follows normal distribution, any data that are extermely 
       skewed are considered as outlier. 
     - identify the outlier of each feature with numeric value.
     - update the outlier values with respective mean observed from training dataset.
    Parameters:
      sigma: number of sigma to detect outlier by default 3 sigma (standard deviation) from mean is considered as outlier
    """

    in_num_sample   = ar_values.shape[0]
    in_ubound       = int(fo_train_ratio * in_num_sample)
    print("INFO: sigma for outliers: {}".format(sigma))

    # find outlier
    print(ar_idxes_int.shape)
    for idx in np.arange(ar_idxes_int.shape[0]):
      fo_mean   = ar_values[:in_ubound, idx].mean(axis=0)
      fo_std    = ar_values[:in_ubound, idx].std(axis=0)
      print("INFO: feature: {0:s} mean: {1:.2f} std: {2:.2f}".format(ts_columns_int[idx], fo_mean, fo_std))

      # Upper bound
      fo_ubound = fo_mean + (sigma * fo_std)
      ar_outlier = (ar_values[:, idx] >= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values[ar_outlier, idx] = fo_mean
        print("INFO: UPPER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, in_num_outlier))

      # Lower bound
      fo_ubound = fo_mean - (sigma * fo_std)
      ar_outlier = (ar_values[:, idx] <= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values[ar_outlier, idx] = fo_mean
        print("INFO: LOWER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, ar_outlier.sum()))

    print("INFO: outliers were identified and updated with respective mean, if required.")
````

### Training and testing data preparation

* Training and testing dataset were created by taking moving average from 100 prior samples. Their corresponding target value is temperature of 24 hours later. Step size of six was used to avoid redundancy. These values such as 100, 24, and 6 can be adjusted but model should be built accordingly.

````python
def make_dataset(in_back_offset=100, in_forward_offset=72, in_step=6, in_min_idx=0, in_max_idx=120000, random=False):
    ''' Aims: moving average 

        Params:
            in_min_idx: lower index
            in_max_idx: upper index
            random: shuffle
    '''
    if in_max_idx is None: in_max_idx = len(ar_values) - 1 # - self._in_forward_offset - 1
    in_idx = in_min_idx + in_back_offset

    # in_b_dim    = in_back_offset//in_step
    in_rows     = (in_max_idx - in_forward_offset - in_idx) // in_step

    in_axis2    = ar_idxes_int.shape[0] + ar_idxes_str.shape[0]
    ar_samples  = np.zeros((in_rows, in_axis2))
    ar_targets  = np.zeros((in_rows, 1))
    
    if random:
      ar_rnd_idx = np.random.randint(in_idx, in_max_idx, in_rows)
      for i in range(ar_rnd_idx.shape[0]):
        in_idx          = ar_rnd_idx[i]
        indices         = np.arange(in_idx-in_back_offset, ar_rnd_idx[i], in_step)
        ar_samples[i, ar_idxes_int]   = np.mean(ar_values[indices[:, None], ar_idxes_int], axis=0)
        if ar_idxes_str.shape[0] > 0:
          ar_samples[i, ts_str_idxes]   = np.median(ar_values[indices[:, None], ar_idxes_str], axis=0).astype(int)
        ar_targets[i]   = ar_values[in_idx + in_forward_offset, in_target_idx]
    else:
      # after preprocessing the data, dimension of new array 
      in_j = 0
      while in_j < ar_samples.shape[0]:
        indices          = np.arange(in_idx-in_back_offset, in_idx, in_step)
        ar_samples[in_j, ar_idxes_int] = np.mean(ar_values[indices[:, None], ar_idxes_int], axis=0)
        if ar_idxes_str.shape[0] > 0:
          ar_samples[in_j, ts_str_idxes]   = np.median(ar_values[indices[:, None], ar_idxes_str], axis=0).astype(int)
        ar_targets[in_j] = ar_values[in_idx + in_forward_offset, in_target_idx]
        in_idx          += in_step
        in_j            += 1
    return ar_samples, ar_targets.ravel()
````

````python
in_train_num = int(fo_train_ratio * ar_values.shape[0])
in_test_num  = int(fo_test_ratio * ar_values.shape[0])
print("# of total data {} - train {} and test {}".format(ar_values.shape[0], in_train_num, in_test_num))

ar_trainX, ar_trainY = make_dataset(100, 72, 6, 0, in_train_num, True)
ar_testX, ar_testY = make_dataset(100, 72, 6, in_train_num+1, None)
print("Training: {} and testing: {}".format(ar_trainX.shape[0], ar_testX.shape[0]))
````

### Manage missing and outlier data

````python
def identify_outlier(sigma = 3.0):
    """
    Aims:
     - assuming each attribute follows normal distribution, any data that are extermely 
       skewed are considered as outlier and are removed. 
     - identify the outlier of each feature with numeric value.
     - update the outlier values with respective mean observed from training dataset.
    Parameters:
      sigma: number of sigma used to calculate outlier 
             by default 3std from mean is considered as outlier
    """

    in_num_sample   = ar_values_wt_date.shape[0]
    in_ubound       = int(fo_train_ratio * in_num_sample)
    print("INFO: sigma for outliers: {}".format(sigma))

    # find outlier
    for idx in np.arange(ar_idxs_wt_date_int.shape[0]):
      fo_mean   = ar_values_wt_date[:in_ubound, idx].mean(axis=0)
      fo_std    = ar_values_wt_date[:in_ubound, idx].std(axis=0)
      print("INFO: feature: {0:s} mean: {1:.2f} std: {2:.2f}".format(ts_cols_wt_date_int[idx], fo_mean, fo_std))

      # Upper bound
      fo_ubound = fo_mean + (sigma * fo_std)
      ar_outlier = (ar_values_wt_date[:, idx] >= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values_wt_date[ar_outlier, idx] = fo_mean
        print("INFO: UPPER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, in_num_outlier))

      # Lower bound
      fo_ubound = fo_mean - (sigma * fo_std)
      ar_outlier = (ar_values_wt_date[:, idx] <= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values_wt_date[ar_outlier, idx] = fo_mean
        print("INFO: LOWER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, ar_outlier.sum()))

    print("INFO: outliers were identified and updated with respective mean, if required.")
````

### Design training and testing set

````python
def make_dataset(in_back_offset=100, in_forward_offset=72, in_step=6, in_min_idx=0, in_max_idx=120000, random=False):
    ''' Aims: moving average 

        Params:
            in_min_idx: lower index
            in_max_idx: upper index
            random: shuffle
    '''
    if in_max_idx is None: in_max_idx = len(ar_values) - 1 # - self._in_forward_offset - 1
    in_idx = in_min_idx + in_back_offset

    # in_b_dim    = in_back_offset//in_step
    in_rows     = (in_max_idx - in_forward_offset - in_idx) // in_step
    in_axis2    = ar_idxs_wt_date_int.shape[0] + ar_idxs_wt_date_str  .shape[0]
    ar_samples_wt_date  = np.zeros((in_rows, in_axis2))
    ar_targets_wt_date  = np.zeros((in_rows, 1))
    
    if random:
      ar_rnd_idx = np.random.randint(in_idx, in_max_idx, in_rows)
      for i in range(ar_rnd_idx.shape[0]):
        in_idx          = ar_rnd_idx[i]
        indices         = np.arange(in_idx-in_back_offset, ar_rnd_idx[i], in_step)
        ar_samples_wt_date[i, ar_idxs_wt_date_int]   = np.mean(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_int], axis=0)
        if ar_idxs_wt_date_str.shape[0] > 0:
          ar_samples_wt_date[i, ar_idxs_wt_date_str]   = np.median(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_str], axis=0).astype(int)
        ar_targets_wt_date[i]   = ar_values_wt_date[in_idx + in_forward_offset, in_target_idx]
    else:
      # after preprocessing the data, dimension of new array 
      in_j = 0
      while in_j < ar_samples_wt_date.shape[0]:
        indices          = np.arange(in_idx-in_back_offset, in_idx, in_step)
        ar_samples_wt_date[in_j, ar_idxs_wt_date_int] = np.mean(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_int], axis=0)
        if ar_idxs_wt_date_str.shape[0] > 0:
          ar_samples_wt_date[in_j, ar_idxs_wt_date_str]   = np.median(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_str], axis=0).astype(int)
        ar_targets_wt_date[in_j] = ar_values_wt_date[in_idx + in_forward_offset, in_target_idx]
        in_idx          += in_step
        in_j            += 1
    return ar_samples_wt_date, ar_targets_wt_date.ravel()
````

````python
in_train_num = int(fo_train_ratio * ar_values_wt_date.shape[0])
in_test_num  = int(fo_test_ratio * ar_values_wt_date.shape[0])
print("# of total data {} - train {} and test {}".format(ar_values_wt_date.shape[0], in_train_num, in_test_num))

ar_train_w_date_X, ar_train_w_date_Y = make_dataset(100, 72, 6, 0, in_train_num, True)
ar_test_w_date_X, ar_test_w_date_Y = make_dataset(100, 72, 6, in_train_num+1, None)
print("Training: {} and testing: {}".format(ar_train_w_date_X.shape[0], ar_test_w_date_X.shape[0]))
````

# 3. More time-dependent features

## 3.1 Objectives

* Month was converted in which Januray and December are close to each other. Similar strategy was also employed for Time and Day.

## 3.2. Data preparation
### Create new features

````python
def convert_time_to_minute(st_row):
    ar_hr_mi_se = st_row.split(":")
    return int(ar_hr_mi_se[0]) * 60 + int(ar_hr_mi_se[1])
          
def get_xy(st_row, fo_max_value = 360.0):
    ar_hr_mi_se = st_row.split(":")
    fo_hrs = (float(ar_hr_mi_se[0]) + float(ar_hr_mi_se[1])/60.0) / fo_max_value
      
    fo_r = fo_max_value/(2.0*np.pi)
    fo_x = fo_r * np.sin(2.0*np.pi*fo_hrs)
    fo_y = fo_r * np.cos(2.0*np.pi*fo_hrs)
    return pd.Series([fo_x, fo_y])
      
def get_month_xy(st_row, fo_max_value = 12.0):
    fo_month = float(st_row) / fo_max_value
      
    fo_r = fo_max_value/(2.0*np.pi)
    fo_x = fo_r * np.sin(2.0*np.pi*fo_month)
    fo_y = fo_r * np.cos(2.0*np.pi*fo_month)
    return pd.Series([fo_x, fo_y])
    
def get_day_xy(st_row, fo_max_value = 365.0):
    ar_dd_mo_yy = st_row.split("-") # dd/mo/yy
    fo_day = (float(ar_dd_mo_yy[0]) + 31.0*(float(ar_dd_mo_yy[1])-1.0)) / fo_max_value

    fo_r = fo_max_value/(2.0*np.pi)
    fo_x = fo_r * np.sin(2.0*np.pi*fo_day)
    fo_y = fo_r * np.cos(2.0*np.pi*fo_day)
    return pd.Series([fo_x, fo_y])

def encode_month(st_row):
    if st_row == 1 or st_row == 2 or st_row == 12:
        return 1
    if st_row == 3 or st_row == 11:
        return 2
    if st_row == 4 or st_row == 10:
        return 3
    else:
        return 4

def encode_hours(st_row):
    if st_row in range(4):
        return 1
    elif st_row in range(4,8):
        return 2
    elif st_row in range(8,12):
        return 3
    elif st_row in range(12, 17):
        return 4
    elif st_row in range(17, 20):
        return 5
    elif st_row in range(20, 24):
        return 6
````

````python
# Converting time, month, and day into numeric format using similar principle 
df_data_wt_date[['Tx', 'Ty']]   = df_data_wt_date['Date Time'].dt.time.astype(str).apply(lambda row: get_xy(row, 24.0))
df_data_wt_date[['Mx', 'My']]   = df_data_wt_date['Date Time'].dt.month.astype(str).apply(lambda row: get_month_xy(row, 12.0))
df_data_wt_date[['Dx', 'Dy']]   = df_data_wt_date['Date Time'].dt.date.astype(str).apply(lambda row: get_day_xy(row, 372.0))

df_data_wt_date['Txy'] = df_data_wt_date['Tx'] +  df_data_wt_date['Ty']
df_data_wt_date['Mxy'] = df_data_wt_date['Mx'] +  df_data_wt_date['My']
df_data_wt_date['Dxy'] = df_data_wt_date['Dx'] +  df_data_wt_date['Dy']
````

### Handling outlier and missing data

````python
def identify_outlier(sigma = 3.0):
    """
    Aims:
     - assuming each attribute follows normal distribution, any data that are extermely 
       skewed are considered as outlier and are removed. 
     - identify the outlier of each feature with numeric value.
     - update the outlier values with respective mean observed from training dataset.
    Parameters:
      sigma: number of sigma used to calculate outlier 
             by default 3std from mean is considered as outlier
    """

    in_num_sample   = ar_values_wt_date.shape[0]
    in_ubound       = int(fo_train_ratio * in_num_sample)
    print("INFO: sigma for outliers: {}".format(sigma))

    # find outlier
    for idx in np.arange(ar_idxs_wt_date_int.shape[0]):
      fo_mean   = ar_values_wt_date[:in_ubound, idx].mean(axis=0)
      fo_std    = ar_values_wt_date[:in_ubound, idx].std(axis=0)
      print("INFO: feature: {0:s} mean: {1:.2f} std: {2:.2f}".format(ts_cols_wt_date_int[idx], fo_mean, fo_std))

      # Upper bound
      fo_ubound = fo_mean + (sigma * fo_std)
      ar_outlier = (ar_values_wt_date[:, idx] >= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values_wt_date[ar_outlier, idx] = fo_mean
        print("INFO: UPPER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, in_num_outlier))

      # Lower bound
      fo_ubound = fo_mean - (sigma * fo_std)
      ar_outlier = (ar_values_wt_date[:, idx] <= fo_ubound)
      in_num_outlier = ar_outlier.sum()
      if in_num_outlier > 0:
        ar_values_wt_date[ar_outlier, idx] = fo_mean
        print("INFO: LOWER BOUND - threshold: {0:.2f} # of outlier: {1:d}".format(fo_ubound, ar_outlier.sum()))

    print("INFO: outliers were identified and updated with respective mean, if required.")
````

### Split dataset for training and testing

````python
def make_dataset(in_back_offset=100, in_forward_offset=72, in_step=6, in_min_idx=0, in_max_idx=120000, random=False):
    ''' Aims: moving average 

        Params:
            in_min_idx: lower index
            in_max_idx: upper index
            random: shuffle
    '''
    if in_max_idx is None: in_max_idx = len(ar_values) - 1 # - self._in_forward_offset - 1
    in_idx = in_min_idx + in_back_offset

    # in_b_dim    = in_back_offset//in_step
    in_rows     = (in_max_idx - in_forward_offset - in_idx) // in_step
    in_axis2    = ar_idxs_wt_date_int.shape[0] + ar_idxs_wt_date_str  .shape[0]
    ar_samples_wt_date  = np.zeros((in_rows, in_axis2))
    ar_targets_wt_date  = np.zeros((in_rows, 1))
    
    if random:
      ar_rnd_idx = np.random.randint(in_idx, in_max_idx, in_rows)
      for i in range(ar_rnd_idx.shape[0]):
        in_idx          = ar_rnd_idx[i]
        indices         = np.arange(in_idx-in_back_offset, ar_rnd_idx[i], in_step)
        ar_samples_wt_date[i, ar_idxs_wt_date_int]   = np.mean(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_int], axis=0)
        if ar_idxs_wt_date_str.shape[0] > 0:
          ar_samples_wt_date[i, ar_idxs_wt_date_str]   = np.median(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_str], axis=0).astype(int)
        ar_targets_wt_date[i]   = ar_values_wt_date[in_idx + in_forward_offset, in_target_idx]
    else:
      # after preprocessing the data, dimension of new array 
      in_j = 0
      while in_j < ar_samples_wt_date.shape[0]:
        indices          = np.arange(in_idx-in_back_offset, in_idx, in_step)
        ar_samples_wt_date[in_j, ar_idxs_wt_date_int] = np.mean(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_int], axis=0)
        if ar_idxs_wt_date_str.shape[0] > 0:
          ar_samples_wt_date[in_j, ar_idxs_wt_date_str]   = np.median(ar_values_wt_date[indices[:, None], ar_idxs_wt_date_str], axis=0).astype(int)
        ar_targets_wt_date[in_j] = ar_values_wt_date[in_idx + in_forward_offset, in_target_idx]
        in_idx          += in_step
        in_j            += 1
    return ar_samples_wt_date, ar_targets_wt_date.ravel()
````

````python
in_train_num = int(fo_train_ratio * ar_values_wt_date.shape[0])
in_test_num  = int(fo_test_ratio * ar_values_wt_date.shape[0])
print("# of total data {} - train {} and test {}".format(ar_values_wt_date.shape[0], in_train_num, in_test_num))

ar_train_w_date_X, ar_train_w_date_Y = make_dataset(100, 72, 6, 0, in_train_num, True)
ar_test_w_date_X, ar_test_w_date_Y = make_dataset(100, 72, 6, in_train_num+1, None)
print("Training: {} and testing: {}".format(ar_train_w_date_X.shape[0], ar_test_w_date_X.shape[0]))
````
